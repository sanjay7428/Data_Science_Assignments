Question 1: Compare and contrast NLTK and spaCy in terms of features, ease of use, and performance.
- NLTK (Natural Language Toolkit) and spaCy are two widely used Python libraries for Natural Language Processing (NLP), but they differ significantly in features, ease of use, and performance.
  - NLTK is primarily designed for teaching and research, offering a comprehensive collection of linguistic resources such as tokenizers, stemmers, lemmatizers, parsers, and access to numerous corpora like WordNet. It provides fine-grained control over individual NLP tasks, making it highly suitable for academic exploration and prototyping. However, it often requires more manual coding and configuration, which can make implementation time-consuming for real-world applications. In contrast, spaCy is designed for industrial and production-level NLP applications, emphasizing efficiency, speed, and scalability. It offers pre-trained pipelines for tasks such as tokenization, part-of-speech tagging, named entity recognition, dependency parsing, and word vectors, all integrated into a streamlined workflow. spaCy's API is more user-friendly and intuitive for building end-to-end applications, reducing boilerplate code.
  - In terms of performance, spaCy significantly outperforms NLTK because it is implemented in Cython and optimized for large-scale data processing, whereas NLTK is comparatively slower and less memory-efficient. While NLTK excels in educational use and flexibility for experimentation, spaCy is better suited for high-performance, real-time NLP systems in industry. Thus, the choice between them depends on whether the objective is academic research and learning (NLTK) or efficient deployment and scalability (spaCy).

Question 2: What is TextBlob and how does it simplify common NLP tasks like sentiment analysis and translation?

Question 4: Describe the architecture and functioning of a Recurrent Natural Network (RNN).
- A Recurrent Neural Network (RNN) is a type of artificial neural network specifically designed to handle sequential data such as text, speech, or time-series signals. Unlike feedforward neural networks, the RNN has a recurrent connection that allows information to persist by maintaining a hidden state, which acts as memory of previous inputs.
  - The basic architecture consists of an input layer, a hidden layer with recurrent connections, and an output layer. At each time step, the network takes the current input and the previous hidden state as inputs, processes them through weighted connections, and produces a new hidden state and output. Mathematically, the hidden state is computed using an activation function such as tanh or ReLU applied to the weighted sum of the current input and the previous hidden state. This recurrent mechanism enables the network to capture temporal dependencies and contextual information across sequences.
  - During training, RNNs use Backpropagation Through Time (BPTT), where errors are propagated backward across multiple time steps to update the weights. However, standard RNNs suffer from vanishing and exploding gradient problems when handling long sequences, making it difficult to learn long-term dependencies. To address this limitation, advanced variants like Long Short-Term Memory (LSTM) and Gated Recurrent Unit (GRU) networks were developed, which include gating mechanisms to control the flow of information. Overall, RNNs are fundamental models for sequential data processing tasks such as language modeling, machine translation, and speech recognition.

Question 5: What is the key difference between LSTM and GRU networks in NLP applications?
- The key difference between Long Short-Term Memory (LSTM) and Gated Recurrent Unit (GRU) networks lies in their internal architecture and complexity for handling long-term dependencies in sequential data.
  - Long Short-Term Memory (LSTM) contains three gates—input gate, forget gate, and output gate—along with a separate cell state that carries long-term memory across time steps. This structure allows LSTM to carefully control what information to store, forget, and output, making it highly effective in capturing long-range dependencies in NLP tasks such as machine translation and language modeling. In contrast, Gated Recurrent Unit (GRU) simplifies this architecture by combining the input and forget gates into a single update gate and merging the cell state with the hidden state.
  - GRU has fewer parameters and a more streamlined structure, leading to faster training and lower computational cost while still maintaining strong performance. While LSTMs may perform slightly better in very complex tasks requiring detailed memory control, GRUs are often preferred in practical NLP applications due to their efficiency and comparable accuracy. Thus, the primary distinction lies in structural complexity, number of gates, and computational efficiency.
  - GRU has fewer parameters and a more streamlined structure, leading to faster training and lower computational cost while still maintaining strong performance. While LSTMs may perform slightly better in very complex tasks requiring detailed memory control, GRUs are often preferred in practical NLP applications due to their efficiency and comparable accuracy.

```python
'''
Question 6: Write a Python program using TextBlob to perform sentiment analysis on
the following paragraph of text:

“I had a great experience using the new mobile banking app. The interface is intuitive,
and customer support was quick to resolve my issue. However, the app did crash once
during a transaction, which was frustrating"

Your program should print out the polarity and subjectivity scores.

'''

from textblob import TextBlob

# Given paragraph
text = """I had a great experience using the new mobile banking app.
The interface is intuitive, and customer support was quick to resolve my issue.
However, the app did crash once during a transaction, which was frustrating."""

# Create TextBlob object
blob = TextBlob(text)

# Get sentiment scores
polarity = blob.sentiment.polarity
subjectivity = blob.sentiment.subjectivity

print("Polarity Score:", polarity)
print("Subjectivity Score:", subjectivity)
```

```python
'''
Question 7: Given the sample paragraph below, perform string tokenization and
frequency distribution using Python and NLTK:

“Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical.”


'''

import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import string

nltk.download('punkt_tab')

# Given paragraph
text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical."""

# Tokenization
text = text.lower()
tokens = word_tokenize(text)

# Remove punctuation
tokens = [word for word in tokens if word.isalpha()]

# Frequency Distribution
freq_dist = FreqDist(tokens)

# Print tokens
print("Tokens:")
print(tokens)

# Print frequency distribution
print("\nWord Frequency Distribution:")
for word, frequency in freq_dist.items():
    print(f"{word}: {frequency}")
```

```python
'''

Question 8: Implement a basic LSTM model in Keras for a text classification task using
the following dummy dataset. Your model should classify sentences as either positive
(1) or negative (0).

# Dataset
texts = [
“I love this project”, #Positive
“This is an amazing experience”, #Positive
“I hate waiting in line”, #Negative
“This is the worst service”, #Negative
“Absolutely fantastic!” #Positive
]
labels = [1, 1, 0, 0, 1]
Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on
this data. You may use Keras with TensorFlow backend.

'''

import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

texts = [
    "I love this project",                 # Positive
    "This is an amazing experience",       # Positive
    "I hate waiting in line",              # Negative
    "This is the worst service",           # Negative
    "Absolutely fantastic!"                # Positive
]

labels = [1, 1, 0, 0, 1]

vocab_size = 1000
max_length = 6
embedding_dim = 16

tokenizer = Tokenizer(num_words=vocab_size, oov_token="<OOV>")
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')

labels = np.array(labels)

# Build LSTM Model
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length),
    LSTM(32),
    Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])


# Train Model

model.fit(padded_sequences, labels, epochs=20, verbose=1)

# Test Prediction

test_sentences = ["I really love this service", "This is worst line."]
test_seq = tokenizer.texts_to_sequences(test_sentences)
test_pad = pad_sequences(test_seq, maxlen=max_length, padding='post')

predictions = model.predict(test_pad)

for sentence, pred in zip(test_sentences, predictions):
    print(f"Sentence: {sentence}")
    print(f"Predicted Sentiment: {'Positive' if pred > 0.5 else 'Negative'}\n")
```

```python
'''
Question 9: Using spaCy, build a simple NLP pipeline that includes tokenization,
lemmatization, and entity recognition. Use the following paragraph as your dataset:

“Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of India’s atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India.”

Write a Python program that processes this text using spaCy, then prints tokens, their
lemmas, and any named entities found.


'''

import spacy

nlp = spacy.load("en_core_web_sm")

# Given paragraph
text = """Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of India’s atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India."""

# Tokenization & Lemmatization

doc = nlp(text)
print("Tokens and Lemmas:\n")
for token in doc:
    print(f"Token: {token.text}  -->  Lemma: {token.lemma_}")


# Named Entity Recognition

print("\nNamed Entities:\n")
for ent in doc.ents:
    print(f"Entity: {ent.text}  |  Label: {ent.label_}")
```
