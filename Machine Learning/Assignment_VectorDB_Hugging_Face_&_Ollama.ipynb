Question 1: What is a Vector Database (VectorDB) and how is it different from traditional databases?
- A Vector Database (VectorDB) is a type of database designed to store, index, and search high-dimensional vector embeddings generated from data such as text, images, audio, or video using machine learning models. Instead of storing data in rows and columns like traditional databases, VectorDBs store numerical vector representations and perform similarity searches using distance metrics such as cosine similarity, Euclidean distance, or dot product.

- Traditional databases like SQL databases store structured data and retrieve information using exact matches or conditional queries (e.g., WHERE clauses). In contrast, VectorDBs are optimized for semantic search, meaning they retrieve data based on similarity in meaning rather than exact keyword matches. This makes VectorDBs especially useful in AI applications such as recommendation systems, semantic search, chatbots, document retrieval, and image search.

Question 2: Explain the various types of VectorDBs available and describe their suitability for different use cases.
- There are several types of Vector Databases, each suited for different use cases based on scalability, performance, and deployment requirements.

1. In-Memory Vector Databases :
These databases store vectors in RAM, making them very fast but less suitable for very large datasets. They are ideal for small to medium AI applications, prototypes, and research projects. Example: Chroma DB.

2. Disk-Based Vector Databases :
These store vectors on disk and are suitable for large datasets where memory storage is not feasible. They are commonly used in enterprise applications and large-scale semantic search systems.

3. Distributed Vector Databases:
These databases distribute data across multiple servers for scalability and high availability. They are suitable for large AI systems, recommendation engines, and large-scale search platforms. Examples include Pinecone, Weaviate, and Milvus.

4. Hybrid Databases :
These combine traditional database features with vector search capabilities. They are useful when both structured data and semantic search are required. Examples include PostgreSQL with pgvector and Elasticsearch with vector search.

Each type is suitable for different scenarios such as research projects, production AI systems, recommendation engines, document search, and retrieval-augmented generation (RAG) systems.

```python
'''
Question 6: Install and set up Chroma DB, and insert sample vector data for semantic
search.

'''
#installing dependencies
!pip install chromadb sentence-transformers

import chromadb
from sentence_transformers import SentenceTransformer
from IPython.display import clear_output

#clearing installation part
clear_output()

#creating main funtion
def main():

  # Initialize Chroma Client
  print("="*50)
  print("Initializing Chroma DB...")
  client = chromadb.Client()

  # Create Collection
  collection = client.create_collection(name="documents")

  # Load Embedding Model
  print("="*50)
  print("Loading embedding model...")
  model = SentenceTransformer('all-MiniLM-L6-v2')

  #Sample Documents
  documents = [
      "Artificial Intelligence is the future",
      "Machine learning is a subset of AI",
      "Deep learning is used in neural networks",
      "Chroma DB is used for vector storage",
      "Semantic search finds similar meaning text",
      "Databases store structured information"
  ]
  #Convert Text to Embeddings
  print("="*50)
  print("Creating embeddings...")
  embeddings = model.encode(documents)

  # Insert Data into Chroma DB
  print("="*50)
  print("Inserting documents into Chroma DB...")
  collection.add(
      documents=documents,
      embeddings=embeddings,
      ids=["id1", "id2", "id3", "id4", "id5", "id6"]
  )

  #Perform Semantic Search
  print("="*50)
  print("\nPerforming Semantic Search...")
  query = "What is Artificial Intelligence?"
  query_embedding = model.encode([query])
  results = collection.query(
      query_embeddings=query_embedding,
      n_results=2
  )

  #Display Results
  print("="*50)
  print("\nQuery:", query)
  print("\nTop Matching Documents:")
  for doc in results['documents'][0]:
      print("-", doc)
  print("="*50)

if __name__ == "__main__":
    main()
```

```python
'''
Question 7: Demonstrate how to download and fine-tune a Hugging Face model for a
text generation task

'''
#IMPORT LIBRARIES
#!pip install transformers datasets torch accelerate

from transformers import AutoTokenizer, AutoModelForCausalLM
from transformers import TrainingArguments, Trainer
from datasets import Dataset
import torch

def main():
    model_name = "gpt2"

    print("Loading model and tokenizer...")
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(model_name)

    tokenizer.pad_token = tokenizer.eos_token
    model.config.pad_token_id = tokenizer.eos_token_id

    # Small dataset
    texts = [
        "Artificial Intelligence is transforming the world.",
        "Machine learning is a subset of artificial intelligence.",
        "Deep learning uses neural networks.",
        "Natural language processing helps computers understand text.",
        "Hugging Face provides many pretrained models.",
        "Transformers are powerful models for text generation.",
        "AI is used in healthcare, finance, and education.",
        "Chatbots use language models to talk to users."
    ]

    dataset = Dataset.from_dict({"text": texts})

    def tokenize_function(examples):
        tokens = tokenizer(
            examples["text"],
            truncation=True,
            padding="max_length",
            max_length=64
        )
        tokens["labels"] = tokens["input_ids"].copy()
        return tokens

    print("Tokenizing dataset...")
    tokenized_dataset = dataset.map(tokenize_function, batched=True)

    training_args = TrainingArguments(
        output_dir="./results",
        num_train_epochs=5,
        per_device_train_batch_size=2,
        logging_steps=10
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=tokenized_dataset
    )

    print("Training model...")
    trainer.train()

    print("Saving model...")
    model.save_pretrained("./fine_tuned_gpt2")
    tokenizer.save_pretrained("./fine_tuned_gpt2")

    # Better text generation
    print("\nGenerating text...")
    input_text = "Artificial Intelligence is"
    inputs = tokenizer(input_text, return_tensors="pt")

    output = model.generate(
        inputs["input_ids"],
        attention_mask=inputs["attention_mask"],
        max_length=50,
        do_sample=True,
        temperature=0.7,
        top_k=50,
        top_p=0.9
    )

    print(tokenizer.decode(output[0], skip_special_tokens=True))


if __name__ == "__main__":
    main()
```

```python
'''

Question 9: Implement a basic RAG (Retrieval-Augmented Generation) system using
Ollama with Llama3.


'''
#Install dependencies
!apt-get update && apt-get install -y zstd
!curl -fsSL https://ollama.com/install.sh | sh
!pip install chromadb requests sentence-transformers

import subprocess
import time
import requests
import chromadb
from sentence_transformers import SentenceTransformer

# Start Ollama Server
def start_ollama():
    print("Starting Ollama server...")
    process = subprocess.Popen(["ollama", "serve"])
    time.sleep(5)
    print("Ollama started!")

# Pull Llama3 Model
def pull_model():
    print("Pulling Llama3 model...")
    subprocess.run(["ollama", "pull", "llama3"])

# Initialize Vector DB
def setup_vectordb():
    client = chromadb.Client()
    collection = client.create_collection(name="docs")
    return collection

#Insert Documents
def add_documents(collection, model):
    docs = [
        "Artificial Intelligence is the simulation of human intelligence.",
        "Machine learning is a subset of AI.",
        "Deep learning uses neural networks.",
        "RAG combines retrieval and generation.",
        "Ollama allows running LLMs locally."
    ]

    embeddings = model.encode(docs)

    collection.add(
        documents=docs,
        embeddings=embeddings,
        ids=[f"id{i}" for i in range(len(docs))]
    )


# Retrieve Relevant Docs
def retrieve(collection, model, query):
    query_embedding = model.encode([query])

    results = collection.query(
        query_embeddings=query_embedding,
        n_results=2
    )

    return results["documents"][0]

# Generate Answer (RAG)
def generate_answer(query, context):
    prompt = f"""
    Use the following context to answer the question.

    Context:
    {context}

    Question:
    {query}
    """

    response = requests.post(
        "http://localhost:11434/api/generate",
        json={
            "model": "llama3",
            "prompt": prompt,
            "stream": False
        }
    )

    return response.json()["response"]

# Main Function
def main():
    start_ollama()
    pull_model()

    print("Loading embedding model...")
    embed_model = SentenceTransformer('all-MiniLM-L6-v2')

    collection = setup_vectordb()
    add_documents(collection, embed_model)

    print("\nRAG System Ready! Type 'exit' to stop.\n")

    while True:
        query = input("You: ")
        if query.lower() == "exit":
            break

        context_docs = retrieve(collection, embed_model, query)
        context = " ".join(context_docs)

        answer = generate_answer(query, context)
        print("AI:", answer)
        print()

if __name__ == "__main__":
    main()
```

Question 10: A health-tech startup wants to build a chatbot that can answer user queries based on medical research articles. Propose and explain a solution using Hugging Face models for understanding, VectorDB for retrieval, and Ollama for generation.

- A health-tech startup can build a medical research chatbot by combining Hugging Face models for text understanding, a Vector Database for document retrieval, and Ollama for response generation in a Retrieval-Augmented Generation (RAG) architecture.
  - First, medical research articles (PDFs, journals, clinical papers) are collected and preprocessed by splitting them into smaller text chunks.
  - These chunks are then converted into embeddings using a Hugging Face sentence transformer model such as sentence-transformers, which captures the semantic meaning of the text.
  - The generated embeddings are stored in a vector database such as Chroma or FAISS.
  - When a user asks a medical question, the query is also converted into an embedding using the same Hugging Face model, and the VectorDB performs similarity search to retrieve the most relevant research text chunks.
  - These retrieved passages are then sent as context along with the user's question to a local large language model running via Ollama (for example, Llama 3 or Mistral), which generates a final, context-aware answer.
  
This approach ensures the chatbot answers based on real medical research rather than general knowledge, reduces hallucinations, maintains data privacy since models can run locally, and provides accurate, research-based responses to users.
