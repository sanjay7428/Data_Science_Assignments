Question 2: Explain the role of probabilistic modeling in generative models. How do these models differ from discriminative models?

- Probabilistic modeling plays a fundamental role in generative models by enabling them to learn the underlying probability distribution of a dataset and generate new data samples that resemble the original data. In probabilistic modeling, a generative model attempts to estimate the joint probability distribution
P(X,Y), where X represents the input data and Y represents the corresponding labels or features. By learning this distribution, the model can understand how the data is generated and can create new synthetic examples that follow the same statistical patterns. Techniques such as Bayesian networks, Hidden Markov Models, Variational Autoencoders (VAEs), and Generative Adversarial Networks (GANs) rely on probabilistic principles to model uncertainty and variability in data. This capability allows generative models to simulate realistic outputs, fill missing data, perform data augmentation, and generate entirely new content such as text, images, or audio.

- Generative models differ from discriminative models in their learning objective and the type of probability distribution they model. Generative models focus on modeling the joint probability
P(X,Y) and can generate new data instances by sampling from this learned distribution. In contrast, discriminative models focus only on learning the conditional probability P(Y|X), which directly maps inputs to labels and is mainly used for prediction or classification tasks. For example, models like Naive Bayes and Gaussian Mixture Models are generative because they attempt to model how the data is generated, while models such as Logistic Regression, Support Vector Machines, and Neural Networks are discriminative because they focus on distinguishing between classes rather than generating new data. Therefore, generative models are more suitable for tasks involving data generation and simulation, whereas discriminative models are typically preferred for tasks such as classification and regression where prediction accuracy is the main goal.

Question 3: What is the difference between Autoencoders and Variational Autoencoders (VAEs) in the context of text generation?

- Autoencoders and Variational Autoencoders (VAEs) are neural network models used to learn efficient representations of data, but they differ significantly in how they represent and generate data, especially in text generation tasks.
  - A standard Autoencoder consists of two main components: an encoder and a decoder. The encoder compresses the input data, such as a sentence or text sequence, into a fixed-length latent vector, while the decoder reconstructs the original input from this compressed representation. The main goal of an Autoencoder is to minimize reconstruction error so that the output is as close as possible to the input. However, the latent space learned by a traditional Autoencoder is deterministic and not structured, which makes it difficult to generate new meaningful text samples because random points in the latent space may not correspond to valid outputs.

  - Variational Autoencoders (VAEs), on the other hand, introduce a probabilistic approach to the latent space. Instead of mapping input text to a single fixed vector, the encoder maps it to a probability distribution defined by parameters such as mean and variance. During training, the model samples latent vectors from this distribution and passes them to the decoder to generate text. VAEs also include a regularization term, usually the Kullback–Leibler (KL) divergence, which ensures that the learned latent space follows a known distribution, typically a normal distribution. This structured latent space allows VAEs to generate new and diverse text samples by sampling from the distribution.

Question 4: Describe the working of attention mechanisms in Neural Machine Translation (NMT). Why are they critical?

- Attention mechanisms in Neural Machine Translation (NMT) improve the translation process by allowing the model to focus on the most relevant parts of the input sentence while generating each word of the output sentence. In traditional encoder–decoder models, the encoder processes the entire source sentence and compresses it into a single fixed-length context vector, which is then passed to the decoder to generate the translated sentence. However, this approach often loses important information, especially when dealing with long sentences. The attention mechanism addresses this limitation by enabling the decoder to access all the hidden states produced by the encoder instead of relying on a single context vector. During translation, for each word the decoder generates, it calculates attention scores that measure how relevant each word in the input sentence is to the current output word. These scores are converted into attention weights using a softmax function, and a weighted sum of the encoder hidden states is computed to form a context vector that highlights the most important information from the input sequence. This context vector is then combined with the decoder’s current state to predict the next word in the translated sentence.

- Attention mechanisms are critical in NMT because they significantly improve translation quality and model performance, especially for long and complex sentences. By dynamically focusing on relevant words in the source sentence, the model can better capture relationships between words and maintain the correct context during translation. This leads to more accurate word alignment between source and target languages and reduces the information bottleneck present in basic encoder–decoder architectures. Additionally, attention improves interpretability because the attention weights can show which parts of the input sentence the model is focusing on while generating each output word. Due to these advantages, attention mechanisms have become a core component of modern translation systems and have also led to the development of advanced architectures such as the Transformer model.

```python
'''
Question 6: Use the following small text dataset to train a simple Variational
Autoencoder (VAE) for text reconstruction:

["The sky is blue", "The sun is bright", "The grass is green",
"The night is dark", "The stars are shining"]

1. Preprocess the data (tokenize and pad the sequences).
2. Build a basic VAE model for text reconstruction.
3. Train the model and show how it reconstructs or generates similar sentences.

Include your code, explanation, and sample outputs.


'''
#------------------<Importing Libraries>----------------------------------------
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, Model, losses
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from IPython.display import clear_output

# ------------------< Dataset >-------------------------------------------------
data = ["The sky is blue", "The sun is bright", "The grass is green",
        "The night is dark", "The stars are shining"]

# -------------------< Tokenization >-------------------------------------------
tokenizer = Tokenizer()
tokenizer.fit_on_texts(data)
sequences = tokenizer.texts_to_sequences(data)

# ---------------------< Padding >----------------------------------------------
max_len = max([len(seq) for seq in sequences])
vocab_size = len(tokenizer.word_index) + 1
x_train = pad_sequences(sequences, maxlen=max_len, padding='post')


class Sampling(layers.Layer):
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.random.normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon

#-----------------------< Hyperparameters >-------------------------------------
latent_dim = 8
embed_dim = 16

#------------------------< Encoder >--------------------------------------------
encoder_inputs = layers.Input(shape=(max_len,))
x = layers.Embedding(vocab_size, embed_dim)(encoder_inputs)
x = layers.LSTM(16)(x)
z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)
z = Sampling()([z_mean, z_log_var])
encoder = Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")

# ----------------------< Decoder >---------------------------------------------
latent_inputs = layers.Input(shape=(latent_dim,))
x = layers.RepeatVector(max_len)(latent_inputs)
x = layers.LSTM(16, return_sequences=True)(x)
decoder_outputs = layers.TimeDistributed(layers.Dense(vocab_size, activation="softmax"))(x)
decoder = Model(latent_inputs, decoder_outputs, name="decoder")


class VAE(Model):
    def __init__(self, encoder, decoder, **kwargs):
        super(VAE, self).__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder

    def train_step(self, data):
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)

            # Reconstruction Loss (Cross-entropy)
            recon_loss = tf.reduce_mean(
    tf.reduce_sum(losses.sparse_categorical_crossentropy(data, reconstruction), axis=1)
)
            # KL Divergence Loss
            kl_loss = -0.5 * tf.reduce_mean(1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
            total_loss = recon_loss + kl_loss

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        return {"loss": total_loss, "recon_loss": recon_loss, "kl_loss": kl_loss}

#-------------------< Compile and Train >---------------------------------------
vae = VAE(encoder, decoder)
vae.compile(optimizer='adam')
vae.fit(x_train, epochs=600, verbose=0)

#----------< Reverse dictionary to decode tokens >------------------------------
index_word = {v: k for k, v in tokenizer.word_index.items()}

def decode_sequence(seq):
    words = []
    for i in seq:
        if i != 0:
            words.append(index_word.get(i, ""))
    return " ".join(words)

#-------------------< Encode and reconstruct sentences >------------------------
z_mean, z_log_var, z = encoder.predict(x_train)
reconstructed = decoder.predict(z)

# ------------------<Convert probabilities to token ids >-----------------------
predicted_sequences = np.argmax(reconstructed, axis=-1)
clear_output()
print("\nSentence Reconstruction:\n")

for i in range(len(data)):
    print("Original:      ", data[i])
    print("Reconstructed: ", decode_sequence(predicted_sequences[i]))
    print()
```

```python
'''
Question 7: Use a pre-trained GPT model (like GPT-2 or GPT-3) to translate a short
English paragraph into French and German. Provide the original and translated text.


'''

#-------------------< Import library >------------------------------------------
from transformers import T5Tokenizer, T5ForConditionalGeneration
from IPython.display import clear_output

#-------------------< Load a model >--------------------------------------------
model_name = "t5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name, model_max_length=512)
model = T5ForConditionalGeneration.from_pretrained(model_name)

#-------------------< translate func >------------------------------------------
def translate_with_t5(text, target_lang):
    input_text = f"translate English to {target_lang}: {text}"

    # Encode and Generate
    input_ids = tokenizer.encode(
        input_text,
        return_tensors="pt"
        )

    outputs = model.generate(
        input_ids,
        max_length=100,
        num_beams=4,
        early_stopping=True
        )

    # Decode
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# ----------------------< Test >------------------------------------------------
source_text = "Completing PWSKILLS assignment text generation and machine translation."

clear_output()

print(f"\nOriginal: {source_text}")
print(f"\nFrench: {translate_with_t5(source_text, 'French')}")
print(f"\nGerman: {translate_with_t5(source_text, 'German')}")
```

```python
'''
Question 8: Implement a simple attention-based encoder-decoder model for
English-to-Spanish translation using Tensorflow or PyTorch.


'''

#-------------< Importing Libraries >-------------------------------------------
import torch
import torch.nn as nn
import torch.nn.functional as F

# ------------< Sample English-Spanish Dataset >--------------------------------

pairs = [
    ("i am happy", "yo estoy feliz"),
    ("i am sad", "yo estoy triste"),
    ("he is tall", "el es alto"),
    ("she is kind", "ella es amable"),
]

# --------------< Build Vocabulary >--------------------------------------------

def build_vocab(sentences):
    vocab = {"<pad>":0,"<sos>":1,"<eos>":2}
    idx = 3

    for sentence in sentences:
        for word in sentence.split():
            if word not in vocab:
                vocab[word] = idx
                idx += 1
    return vocab


eng_sentences = [p[0] for p in pairs]
spa_sentences = [p[1] for p in pairs]

src_vocab = build_vocab(eng_sentences)
trg_vocab = build_vocab(spa_sentences)

src_vocab_size = len(src_vocab)
trg_vocab_size = len(trg_vocab)

# Reverse vocab
inv_trg_vocab = {v:k for k,v in trg_vocab.items()}


# --------------< Encode Sentences >--------------------------------------------

def encode(sentence, vocab):
    tokens = [vocab[word] for word in sentence.split()]
    return torch.tensor([1] + tokens + [2])  # <sos> sentence <eos>

src_data = [encode(p[0], src_vocab) for p in pairs]
trg_data = [encode(p[1], trg_vocab) for p in pairs]


#------------< Attention Mechanism >--------------------------------------------

class BahdanauAttention(nn.Module):

    def __init__(self, hidden_dim):
        super().__init__()

        self.W1 = nn.Linear(hidden_dim, hidden_dim)
        self.W2 = nn.Linear(hidden_dim, hidden_dim)
        self.V = nn.Linear(hidden_dim,1)

    def forward(self, query, values):

        score = self.V(torch.tanh(self.W1(query) + self.W2(values)))
        attention_weights = F.softmax(score, dim=1)

        context_vector = torch.sum(attention_weights * values, dim=1)

        return context_vector, attention_weights

#----------------< Encoder >----------------------------------------------------

class Encoder(nn.Module):

    def __init__(self, vocab_size, embed_dim, hidden_dim):
        super().__init__()

        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.gru = nn.GRU(embed_dim, hidden_dim, batch_first=True)

    def forward(self, x):

        embedded = self.embedding(x)
        outputs, hidden = self.gru(embedded)

        return outputs, hidden

#----------------< Decoder >----------------------------------------------------
class Decoder(nn.Module):

    def __init__(self, vocab_size, embed_dim, hidden_dim):
        super().__init__()

        self.attention = BahdanauAttention(hidden_dim)

        self.embedding = nn.Embedding(vocab_size, embed_dim)

        self.gru = nn.GRU(embed_dim + hidden_dim, hidden_dim, batch_first=True)

        self.fc = nn.Linear(hidden_dim, vocab_size)

    def forward(self, x, hidden, encoder_outputs):

        embedded = self.embedding(x)

        context_vector, _ = self.attention(hidden.transpose(0,1), encoder_outputs)

        combined = torch.cat((embedded, context_vector.unsqueeze(1)), dim=2)

        output, hidden = self.gru(combined, hidden)

        prediction = self.fc(output.squeeze(1))

        return prediction, hidden

#---------------< Seq2Seq Model >-----------------------------------------------
class Seq2Seq(nn.Module):

    def __init__(self, encoder, decoder):
        super().__init__()

        self.encoder = encoder
        self.decoder = decoder

    def forward(self, src, trg):

        encoder_outputs, hidden = self.encoder(src)

        outputs = []

        input_token = trg[:,0].unsqueeze(1)

        for t in range(1, trg.shape[1]):

            output, hidden = self.decoder(input_token, hidden, encoder_outputs)

            outputs.append(output)

            input_token = trg[:,t].unsqueeze(1)

        return torch.stack(outputs, dim=1)

#-------------< Initialize Model >----------------------------------------------

embed_dim = 32
hidden_dim = 64

encoder = Encoder(src_vocab_size, embed_dim, hidden_dim)
decoder = Decoder(trg_vocab_size, embed_dim, hidden_dim)

model = Seq2Seq(encoder, decoder)

optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = nn.CrossEntropyLoss()

#---------------< Training >----------------------------------------------------

for epoch in range(500):

    total_loss = 0

    for src, trg in zip(src_data, trg_data):

        src = src.unsqueeze(0)
        trg = trg.unsqueeze(0)

        optimizer.zero_grad()

        output = model(src, trg)

        output = output.reshape(-1, trg_vocab_size)
        target = trg[:,1:].reshape(-1)

        loss = criterion(output, target)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    if epoch % 100 == 0:
        print("Epoch:", epoch, "Loss:", total_loss)

#-----------------< Translation Function >--------------------------------------
def translate(sentence):

    model.eval()

    src = encode(sentence, src_vocab).unsqueeze(0)

    encoder_outputs, hidden = encoder(src)

    input_token = torch.tensor([[1]])

    result = []

    for _ in range(10):

        output, hidden = decoder(input_token, hidden, encoder_outputs)

        pred = output.argmax(1).item()

        if pred == 2:
            break

        result.append(inv_trg_vocab[pred])

        input_token = torch.tensor([[pred]])

    return " ".join(result)

#-------------< Test Translation >----------------------------------------------
print("\nTranslation Results:\n")

print("i am happy ->", translate("i am happy"))
print("i am sad ->", translate("i am sad"))
print("she is kind ->", translate("she is kind"))
```
