Question 1: What are Large Language Models (LLMs) and how do they function?
- Large Language Models (LLMs) are artificial intelligence models trained on massive amounts of text data to understand, generate, summarize, and respond to human language. They are mainly based on the Transformer architecture, which uses attention mechanisms to understand relationships between words in context.

- LLMs function in the following way:
1. Input text is broken into tokens (small pieces of words).
2. Tokens are passed through multiple neural network layers.
3. The model analyzes patterns and relationships between tokens using attention.
4. The model predicts the most probable next token based on probability.
5. Tokens are generated sequentially to form a response.

```python

'''
Question 7: Integrate Langchain with a third-party API (e.g., weather, news) and show
how responses can be generated via LLMs.

'''
#Installing Dependencies

!pip install -U "langchain[google-genai]"

import requests
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.messages import HumanMessage
from IPython.display import clear_output,Markdown
from google.colab import userdata

#Lets clear the output screen
clear_output()

#Ask the location for weather api
Place = input ("Enter your current location : ")
print("\n\n")

# API Keys for colab secret
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
WEATHER_API_KEY = userdata.get('Weatherapi')

#Get weather data from Weather API

url = f"http://api.weatherapi.com/v1/current.json?key={WEATHER_API_KEY}&q={Place}"
response = requests.get(url)
weather_json = response.json()


# Extract useful data
temp_c = weather_json["current"]["temp_c"]
condition = weather_json["current"]["condition"]["text"]
humidity = weather_json["current"]["humidity"]
wind_kph = weather_json["current"]["wind_kph"]


#Create Prompt Template
template = """
You are a fashion and weather assistant.

Weather details:
Temperature: {temp} °C
Condition: {condition}
Humidity: {humidity}
Wind Speed: {wind}

Question: What should I wear today?

Give practical clothing advice according to weather.
"""

prompt = PromptTemplate(
    input_variables=["temp", "condition", "humidity", "wind"],
    template=template
)

final_prompt = prompt.format(
    temp=temp_c,
    condition=condition,
    humidity=humidity,
    wind=wind_kph
)

#Creating instance of Gemini Model
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite", google_api_key=GOOGLE_API_KEY)

#Get Response
response = model.invoke(
    [HumanMessage(content=final_prompt)]
)

#Result time
display(Markdown(response.content))
```

```python
'''
Question 9: Demonstrate combining Langchain with LamaIndex to create a simple
document-based Q&A chatbot.


'''

#Install Dependencies

!pip install llama-index
!pip install llama-index-llms-google-genai
!pip install llama-index-embeddings-huggingface
!pip install llama-index-readers-file
!pip install langchain
!pip install -U "langchain[google-genai]"

import os

# LlamaIndex imports
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

# LangChain imports
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate

# other imports
from IPython.display import clear_output,Markdown
from google.colab import userdata
clear_output()

# Set API Key
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')

# Setup Embedding Model
Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en")

# Load and Index Documents
documents = SimpleDirectoryReader("/content/mydata").load_data()
index = VectorStoreIndex.from_documents(documents)

# Create retriever
retriever = index.as_retriever()

# Setup Gemini
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite",api_key=GOOGLE_API_KEY)

# Prompt Template
prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
You are a helpful assistant.

Use the context below to answer the question.

Context:
{context}

Question:
{question}
"""
)

clear_output()

# Chat Loop
while True:
  question = input("Ask your question (type 'exit' to quit): ")

  if question.lower() == "exit":
    print("Exiting chatbot...")
    break

  nodes = retriever.retrieve(question)
  context = "\n".join([node.text for node in nodes])

  final_prompt = prompt.format(
        context=context,
        question=question
    )

  response = llm.invoke(final_prompt)
  print("\nAnswer:", response.content,"\n")
  print("-"*20)
```
