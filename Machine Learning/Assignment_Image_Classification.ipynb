Question 1: What is a Convolutional Neural Network (CNN), and how does it differ from traditional fully connected neural networks in terms of architecture and performance on image data?
- A Convolutional Neural Network (CNN) is a deep learning architecture specifically designed to process image and spatial data by exploiting the spatial relationships between pixels. Unlike traditional fully connected neural networks (FCNNs), which treat input data as a one-dimensional vector, CNNs preserve the two-dimensional structure of images (height, width, and channels). The core building block of a CNN is the convolutional layer, where small learnable filters (kernels) slide across the input image to extract local features such as edges, corners, and textures. These local features are then combined across multiple layers to learn higher-level representations like shapes and objects. CNNs also include pooling layers, which down-sample feature maps to reduce spatial dimensions, computational cost, and sensitivity to small translations, thereby improving generalization.

- In contrast, traditional fully connected neural networks connect every neuron in one layer to every neuron in the next, resulting in a very large number of parameters when applied to images. This requires flattening the image into a vector, which destroys spatial information and makes it difficult for the network to learn meaningful visual patterns. Due to the lack of local connectivity and weight sharing, FCNNs are computationally expensive, memory-intensive, and highly prone to overfitting, especially for high-resolution images. CNNs overcome these limitations through local receptive fields and weight sharing, significantly reducing the number of trainable parameters while still capturing essential visual patterns. As a result, CNNs achieve better accuracy, faster training, and stronger generalization on image-related tasks such as image classification, object detection, and medical image analysis, making them the preferred architecture for computer vision problems.

Question 2: Discuss the architecture of LeNet-5 and explain how it laid the foundation for modern deep learning models in computer vision. Include references to its original research paper.
- LeNet-5 is one of the earliest and most influential Convolutional Neural Network (CNN) architectures, proposed by Yann LeCun, Léon Bottou, Yoshua Bengio, and Patrick Haffner in their seminal paper “Gradient-Based Learning Applied to Document Recognition” (1998). It was originally designed for handwritten digit recognition, particularly for recognizing ZIP codes and bank cheques, and it laid the conceptual and architectural foundation for modern deep learning models in computer vision.

- The architecture of LeNet-5 follows a layered, hierarchical design that alternates between convolution and subsampling (pooling) layers, followed by fully connected layers. It takes a 32x32 grayscale image as input. The first convolutional layer (C1) applies multiple 5x5 filters to extract low-level features such as edges, producing feature maps. This is followed by a subsampling (pooling) layer (S2), which reduces spatial dimensions and introduces translation invariance. The process is repeated with a second convolutional layer (C3) that learns more complex patterns by combining features from earlier layers, followed again by another subsampling layer (S4). The extracted high-level features are then passed to fully connected layers (C5 and F6), and finally to an output layer that performs classification using a form of softmax. A key innovation in LeNet-5 was the use of weight sharing and local receptive fields, which drastically reduced the number of parameters compared to fully connected networks and made learning from image data feasible.

- LeNet-5 laid the foundation for modern computer vision models in several important ways.
  - First, it demonstrated that end-to-end learning directly from raw pixels using gradient-based optimization is effective for visual tasks.
  - Second, it introduced the now-standard CNN design pattern of convolution => pooling => deeper convolution => fully connected layers, which is still used in architectures like AlexNet, VGG, and ResNet.
  - Third, its use of spatial feature hierarchies showed that simple features learned in early layers can be composed into complex representations in deeper layers.
  
  Although limited by computational power and dataset size at the time, LeNet-5 established the core principles that enabled the deep learning revolution in computer vision decades later, making it a cornerstone model in the history of neural networks.

Question 3: Compare and contrast AlexNet and VGGNet in terms of design principles, number of parameters, and performance. Highlight key innovations and limitations of each.
- AlexNet and VGGNet are two landmark convolutional neural network architectures that significantly advanced the field of computer vision, particularly in large-scale image recognition tasks such as the ImageNet Large Scale Visual Recognition Challenge (ILSVRC). While both models follow the core CNN paradigm of stacked convolutional and pooling layers followed by fully connected layers, they differ notably in their design philosophy, architectural complexity, number of parameters, and overall performance.

- AlexNet :  Introduced by Krizhevsky, Sutskever, and Hinton in 2012, was designed to demonstrate that deep CNNs could be effectively trained on large datasets using GPUs. Its architecture consists of 8 layers (5 convolutional and 3 fully connected layers) and employs relatively large convolutional filters (such as 11x11 and 5x5) in the early layers to quickly reduce spatial dimensions. AlexNet introduced several key innovations, including the use of ReLU activation functions to mitigate the vanishing gradient problem, dropout in fully connected layers to reduce overfitting, and data augmentation to improve generalization. It also used local response normalization (LRN), inspired by biological neural mechanisms. AlexNet achieved a dramatic improvement in ImageNet performance, winning ILSVRC 2012 by a large margin. However, its limitations include a large number of parameters (around 60 million), heavy reliance on fully connected layers, and the use of large kernels, which are computationally expensive and less efficient in capturing fine-grained features.

- VGGNet : Proposed by Simonyan and Zisserman in 2014, adopted a very different design principle focused on depth and simplicity. Instead of large convolutional kernels, VGGNet uses very small 3x3 filters stacked sequentially, allowing the network to be much deeper (16 or 19 layers) while maintaining a consistent architectural pattern. This design enables the model to learn more complex and hierarchical feature representations while preserving spatial resolution. As a result, VGGNet achieved better accuracy than AlexNet on ImageNet and became a widely used feature extractor in transfer learning. However, this increased depth came at the cost of a significantly larger number of parameters (approximately 138 million), leading to high memory usage, slow training, and expensive inference. Additionally, VGGNet lacks architectural innovations to address vanishing gradients or computational efficiency, which were later solved by models such as ResNet and Inception.

Question 4: What is transfer learning in the context of image classification? Explain how it helps in reducing computational costs and improving model performance with limited data.
- Transfer learning in the context of image classification refers to the technique of leveraging a pre-trained deep learning model, typically trained on a large and diverse dataset such as ImageNet, and adapting it to a new but related image classification task. Instead of training a convolutional neural network (CNN) from scratch, transfer learning reuses the learned weights of earlier layers, which capture generic visual features like edges, textures, shapes, and object parts that are common across many image domains. These pre-trained models (such as VGG, ResNet, Inception, or EfficientNet) serve as strong feature extractors, while only the final layers are modified and retrained to suit the new classification problem.

- Transfer learning significantly reduces computational costs because training a deep CNN from scratch requires massive datasets, long training times, and high-end GPU resources. By starting from pre-trained weights, the model converges much faster, requires fewer training epochs, and avoids the expensive process of learning low-level features again. In many cases, only a small portion of the network—often the final classification layers—is trained, further reducing memory usage and computation. This makes transfer learning particularly practical in real-world scenarios where computational resources are limited.

- Transfer learning improves model performance when data is scarce. With limited labeled images, training a deep network from scratch often leads to severe overfitting and poor generalization. Pre-trained models provide a strong initialization that already encodes rich visual representations, enabling the model to generalize better even with small datasets. Fine-tuning selected deeper layers allows the network to adapt high-level features to the new domain while retaining robust low-level representations. As a result, transfer learning often achieves higher accuracy, better stability, and faster convergence compared to models trained from scratch, making it a widely adopted strategy in medical imaging, satellite imagery, and other data-constrained image classification tasks.

Question 5: Describe the role of residual connections in ResNet architecture. How do they address the vanishing gradient problem in deep CNNs?
- Residual connections, also known as skip connections, are the core innovation of the ResNet (Residual Network) architecture introduced by He et al. (2015) to enable the training of very deep convolutional neural networks. In a traditional deep CNN, each layer is expected to learn a direct mapping from input to output, which becomes increasingly difficult as the network depth grows. Residual connections address this challenge by allowing the input of a layer to be directly added to its output, so that the layer learns only a residual function—that is, the difference between the desired output and the input—rather than the complete transformation.

  Formally, instead of learning a mapping H(x), a residual block learns F(x)=H(x)-x, and the final output becomes y=F(x)+x. This simple yet powerful design has a profound impact on training deep networks. By providing a direct pathway for information and gradients to flow, residual connections ensure that gradients can bypass multiple layers during backpropagation. This prevents gradients from becoming extremely small as they propagate backward through many layers, effectively mitigating the vanishing gradient problem that traditionally limited the depth of CNNs.

  As a result, residual connections make it easier to optimize very deep networks—sometimes exceeding 100 layers—by ensuring that earlier layers continue to receive meaningful gradient signals during training. If deeper layers fail to learn useful transformations, the residual connection allows the network to behave like a shallower model, ensuring that performance does not degrade with increased depth. This innovation enabled ResNet to achieve state-of-the-art performance on ImageNet and laid the foundation for many modern architectures, such as DenseNet and EfficientNet, by demonstrating that depth can be increased without sacrificing trainability or accuracy.

```python
'''
Question 6: Implement the LeNet-5 architectures using Tensorflow or PyTorch to
classify the MNIST dataset. Report the accuracy and training time.

'''

import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
import time

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Reshape and normalize
x_train = x_train.reshape(-1, 28, 28, 1) / 255.0
x_test = x_test.reshape(-1, 28, 28, 1) / 255.0

#LeNet5 architechture
model = models.Sequential([
    layers.Conv2D(6, kernel_size=5, activation='tanh', input_shape=(28, 28, 1)),
    layers.AveragePooling2D(pool_size=(2,2)),

    layers.Conv2D(16, kernel_size=5, activation='tanh'),
    layers.AveragePooling2D(pool_size=(2,2)),

    layers.Flatten(),
    layers.Dense(120, activation='tanh'),
    layers.Dense(84, activation='tanh'),
    layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Training model
start_time = time.time()

history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.1,
    verbose=1
)

training_time = time.time() - start_time

# Eval
test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)

print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Training Time: {training_time:.2f} seconds")
```

```python
'''
Question 8: Write a program to visualize the filters and feature maps of the first
convolutional layer of AlexNet on an example input image.

'''

import torch
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Load pre-trained AlexNet
model = models.alexnet(pretrained=True)
model.eval()

# Extract first convolutional layer weights
first_conv_layer = model.features[0]
filters = first_conv_layer.weight.data.clone()

# Normalize filters for visualization
filters = (filters - filters.min()) / (filters.max() - filters.min())

# Plot filters
fig, axes = plt.subplots(8, 8, figsize=(10, 10))
for i, ax in enumerate(axes.flat):
    filter_img = filters[i].permute(1, 2, 0)
    ax.imshow(filter_img)
    ax.axis('off')

plt.suptitle("AlexNet First Convolutional Layer Filters")
plt.show()

# Image preprocessing (AlexNet expects 224x224 images)
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

img = Image.open("/content/cat1.jpg").convert("RGB")#image path here
input_tensor = transform(img).unsqueeze(0)


# Forward pass through first conv layer only
with torch.no_grad():
    feature_maps = model.features[0](input_tensor)

# Convert to numpy
feature_maps = feature_maps.squeeze(0).cpu().numpy()

fig, axes = plt.subplots(8, 8, figsize=(10, 10))
for i, ax in enumerate(axes.flat):
    ax.imshow(feature_maps[i], cmap='gray')
    ax.axis('off')

plt.suptitle("Feature Maps from First Conv Layer of AlexNet")
plt.show()
```

```python
'''
Question 9: Train a GoogLeNet (Inception v1) or its variant using a standard dataset
like CIFAR-10. Plot the training and validation accuracy over epochs and analyze
overfitting or underfitting.
'''

import tensorflow as tf
import matplotlib.pyplot as plt

# Load dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# Normalize
x_train = x_train / 255.0
x_test = x_test / 255.0

#inception fuction
def inception_module(x, f1, f3, f5, fp):
    conv1 = tf.keras.layers.Conv2D(f1, (1,1), padding='same', activation='relu')(x)

    conv3 = tf.keras.layers.Conv2D(f3, (3,3), padding='same', activation='relu')(x)

    conv5 = tf.keras.layers.Conv2D(f5, (5,5), padding='same', activation='relu')(x)

    pool = tf.keras.layers.MaxPooling2D((3,3), strides=(1,1), padding='same')(x)
    pool = tf.keras.layers.Conv2D(fp, (1,1), padding='same', activation='relu')(pool)

    return tf.keras.layers.Concatenate()([conv1, conv3, conv5, pool])

# Building model

inputs = tf.keras.Input(shape=(32, 32, 3))

x = tf.keras.layers.Conv2D(64, (3,3), padding='same', activation='relu')(inputs)
x = tf.keras.layers.MaxPooling2D((2,2))(x)

x = inception_module(x, 32, 64, 16, 16)
x = inception_module(x, 64, 96, 32, 32)

x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(128, activation='relu')(x)
outputs = tf.keras.layers.Dense(10, activation='softmax')(x)

model = tf.keras.Model(inputs, outputs)

#model compile
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

#model training
history = model.fit(
    x_train, y_train,
    validation_split=0.2,
    epochs=20,
    batch_size=64,
    verbose=1
)

#visualising
plt.figure()
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend(["Training Accuracy", "Validation Accuracy"])
plt.title("Training vs Validation Accuracy (GoogLeNet on CIFAR-10)")
plt.show()
```
