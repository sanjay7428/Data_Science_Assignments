## Theoretical Questions and Answers

1. What is unsupervised learning in the context of machine learning?
- Unsupervised learning is a machine learning approach where the model learns patterns from unlabeled data. Unlike supervised learning, it does not rely on predefined output labels. Instead, it discovers hidden structures, relationships, or groupings within the data. Common unsupervised tasks include clustering, dimensionality reduction, and association rule mining.

---

2. How does K-Means clustering algorithm work?
- K-Means is a centroid-based clustering algorithm that partitions data into K clusters. It begins by initializing K random centroids. Each data point is then assigned to the nearest centroid based on distance (usually Euclidean distance). After assignment, the centroids are recalculated by taking the mean of all points in each cluster. This process of assignment and centroid update repeats until the centroids stabilize or the algorithm reaches a maximum number of iterations.

---

3. Explain the concept of a dendrogram in hierarchical clustering.
- A dendrogram is a tree-like diagram used to visualize the arrangement of clusters formed during hierarchical clustering. It shows the progressive merging (agglomerative) or splitting (divisive) of clusters. The vertical axis represents the distance or similarity between clusters, and the height at which two clusters join indicates how similar or different they are. By cutting the dendrogram at a chosen level, users can decide the number of clusters.

---

4. What is the main difference between K-Means and Hierarchical Clustering?
- The primary difference is that K-Means is a partitioning method requiring the number of clusters (K) beforehand, while hierarchical clustering builds a tree structure of clusters without requiring a preset number. K-Means is efficient for large datasets, whereas hierarchical clustering is more computationally expensive but provides richer information through dendrograms.

---

5. What are the advantages of DBSCAN over K-Means?
- DBSCAN can discover clusters of arbitrary shapes, unlike K-Means which works best on spherical clusters. It also does not require specifying the number of clusters beforehand. Additionally, DBSCAN naturally identifies noise or outliers, which K-Means cannot handle effectively. It is particularly strong when dealing with clusters that vary in density.

---

6. When would you use Silhouette Score in clustering?
- The Silhouette Score is used to evaluate how well data points fit within their assigned clusters. It measures both cohesion (how close points are within a cluster) and separation (how far a cluster is from others). It is typically used when determining the optimal number of clusters or comparing different clustering algorithms to assess quality.

---

7. What are the limitations of Hierarchical Clustering?
- Hierarchical clustering is computationally expensive, especially for large datasets, because it calculates and stores all pairwise distances. It is also sensitive to noise and outliers. Once a merge or split is done, it cannot be undone, making the algorithm less flexible than iterative methods like K-Means.

---

8. Why is feature scaling important in clustering algorithms like K-Means?
- Feature scaling ensures that all variables contribute equally to the distance calculations used in K-Means. Since K-Means relies on Euclidean distance, features with larger magnitude can dominate others, distorting cluster formation. Scaling techniques like standardization or normalization ensure fair comparison across features.

---

9. How does DBSCAN identify noise points?
- DBSCAN classifies a point as noise if it has fewer than the minimum required number of neighboring points within a specified radius (epsilon). Such points do not belong to any cluster because they fail to meet the density criteria needed to form a cluster.

---

10. Define inertia in the context of K-Means.
- Inertia refers to the sum of squared distances between each data point and its assigned cluster centroid. It measures how compact the clusters are. Lower inertia indicates tighter, more cohesive clusters, and it is often used to help determine the optimal number of clusters using the Elbow Method.

---

11. What is the elbow method in K-Means clustering?
- The Elbow Method is a technique used to determine the optimal number of clusters (K) in K-Means. It involves plotting the inertia (within-cluster sum of squared distances) against different values of K. As K increases, inertia decreases, but after a certain point, the rate of decrease slows sharply. This point looks like an elbow on the graph and is considered the ideal number of clusters, balancing cluster quality and simplicity.

---

12. Describe the concept of "density" in DBSCAN.
- In DBSCAN, density refers to how closely points are packed together within a region. A region is considered dense if a point has at least the minimum required number of neighboring points within a specified radius (epsilon). DBSCAN uses this idea to form clusters based on areas of high density while identifying sparse areas as noise.

---

13. Can hierarchical clustering be used on categorical data?
- Yes, hierarchical clustering can be applied to categorical data, but it requires appropriate distance measures such as Hamming distance or matching coefficients. Traditional hierarchical clustering is designed for numeric data, so special preprocessing or mixed-type distance metrics are needed when dealing with categorical or mixed datasets.

---

14. What does a negative Silhouette Score indicate?
- A negative Silhouette Score indicates that a data point is likely assigned to the wrong cluster. It means the point is closer to points in another cluster than to those in its own, suggesting overlapping clusters or poor cluster structure. Negative values signal that the clustering solution may need improvement.

---

15. Explain the term "linkage criteria" in hierarchical clustering.
- Linkage criteria determine how the distance between clusters is calculated when merging them in hierarchical clustering. Common linkage methods include single linkage (minimum distance), complete linkage (maximum distance), average linkage (mean distance), and Ward's method (minimizing variance). The choice of linkage affects cluster shape and structure.

---

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
- K-Means assumes clusters are spherical, equally sized, and evenly dense. When clusters vary in size or density, points may be incorrectly grouped because K-Means solely relies on minimizing distances to centroids. Larger or denser clusters can dominate centroid positions, leading to inaccurate boundaries and poor performance.

---

17. What are the core parameters in DBSCAN, and how do they influence clustering?
- The main parameters in DBSCAN are epsilon (ε) and minPts.
  - ε (epsilon) defines the radius within which neighboring points are considered. A large ε may merge separate clusters, while a small ε may break clusters or label many points as noise.
  - minPts defines the minimum number of points needed to form a dense region. Higher minPts leads to stricter cluster formation, while lower values can cause noise to form small clusters.

Together, these parameters decide cluster size, shape, and noise detection.

---

18. How does K-Means++ improve upon standard K-Means initialization?
- K-Means++ improves cluster initialization by choosing initial centroids more strategically rather than randomly. It starts with one random centroid, then selects subsequent centroids with probability proportional to their distance squared from existing centroids. This spreads centroids apart, reducing the chances of poor initialization and typically leading to faster convergence and better clustering.

---

19. What is agglomerative clustering?
- Agglomerative clustering is a bottom-up hierarchical clustering method. It starts by treating each data point as an individual cluster. Then, step by step, it merges the two closest clusters based on a chosen distance metric and linkage criterion until all points are in a single cluster or a desired number of clusters is reached.

---

20. What makes Silhouette Score a better metric than just inertia for model evaluation?
- Silhouette Score is considered better than inertia because it accounts for both cluster cohesion (how close points are within a cluster) and cluster separation (how far clusters are from each other). Inertia only measures compactness and does not consider the distance between clusters, making it less informative. Silhouette Score provides a more holistic and comparable evaluation across different numbers of clusters.

---

```python
'''

29. Create synthetic data using make_blobs and evaluate silhouette
scores for k = 2 to 5. Display as a bar chart.

'''

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X, _ = make_blobs(n_samples=1000, centers=4, cluster_std=0.75, random_state=42)

scores = {}
for k in range(2, 6):
    model = KMeans(n_clusters=k, random_state=42)
    labels = model.fit_predict(X)
    scores[k] = silhouette_score(X, labels)

plt.figure(figsize=(6, 4))
plt.bar(scores.keys(), scores.values())
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for k = 2 to 5")
plt.tight_layout()
plt.show()
```

```python
'''

30. Load the Iris dataset and use hierarchical clustering to group data.
Plot a dendrogram with average linkage.

'''

from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data

Z = linkage(X, method='average')

plt.figure(figsize=(10, 6))
dendrogram(Z, truncate_mode=None)
plt.title("Hierarchical Clustering Dendrogram (Average Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.tight_layout()
plt.show()
```

```python
'''

31. Generate synthetic data with overlapping clusters using make_blobs,
then apply K-Means and visualize with decision boundaries.

'''
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y = make_blobs(
    n_samples=600,
    centers=3,
    cluster_std=2.0,
    random_state=42
)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

x_min, x_max = X[:,0].min() - 1, X[:,0].max() + 1
y_min, y_max = X[:,1].min() - 1, X[:,1].max() + 1
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 500),
    np.linspace(y_min, y_max, 500)
)

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:,0], X[:,1], c=labels, s=40, edgecolor='k')
plt.scatter(
    kmeans.cluster_centers_[:,0],
    kmeans.cluster_centers_[:,1],
    c='red', s=200, marker='X'
)
plt.title("K-Means Clustering with Decision Boundaries")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.tight_layout()
plt.show()
```

```python
'''

34. Load the Breast Cancer dataset and compare inertia values for
K = 2 to 6 using K-Means. Show results in a line plot.

'''

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans

data = load_breast_cancer()
X = data.data

inertias = []
K_values = range(2, 7)

for k in K_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Plot results
plt.figure(figsize=(7, 5))
plt.plot(K_values, inertias, marker='o')
plt.title("K-Means Inertia for K = 2 to 6")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
'''

43. Generate noisy circular data using make_circles and visualize
clustering results from KMeans and DBSCAN side-by-side.

'''
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler

X, y = make_circles(
    n_samples=600,
    factor=0.5,
    noise=0.07,
    random_state=42
)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, s=40, edgecolor='k')
plt.title("KMeans Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, s=40, edgecolor='k')
plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.tight_layout()
plt.show()
```

```python
'''

44. Load the Iris dataset and plot the Silhouette Coefficient for
each sample after KMeans clustering.

'''

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X = iris.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

silhouette_vals = silhouette_samples(X_scaled, labels)

plt.figure(figsize=(10, 5))
plt.plot(silhouette_vals, marker='o', linewidth=1)
plt.title("Silhouette Coefficient for Each Sample (Iris Dataset)")
plt.xlabel("Sample Index")
plt.ylabel("Silhouette Coefficient")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
'''

45. Generate synthetic data using make_blobs and apply Agglomerative
Clustering with 'average' linkage. Visualize clusters.

'''
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering

X, y = make_blobs(
    n_samples=500,
    centers=4,
    cluster_std=1.2,
    random_state=42
)

agg = AgglomerativeClustering(
    n_clusters=4,
    linkage='average'
)
labels = agg.fit_predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, edgecolor='k')
plt.title("Agglomerative Clustering (Average Linkage)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.tight_layout()
plt.show()
```

```python
'''

46. Load the Wine dataset, apply KMeans, and visualize the cluster
assignments in a seaborn pairplot (first 4 features).

'''

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

wine = load_wine()
X = wine.data[:, :4]
feature_names = wine.feature_names[:4]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

df = pd.DataFrame(X, columns=feature_names)
df["Cluster"] = labels

sns.pairplot(df, hue="Cluster", diag_kind="kde", palette="tab10")
plt.suptitle("KMeans Clustering on Wine Dataset (First 4 Features)", y=1.02)
plt.show()
```

```python
'''

47. Generate noisy blobs using make_blobs and use DBSCAN to identify
both clusters and noise points. Print the count.

'''

from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np

X, y = make_blobs(
    n_samples=700,
    centers=4,
    cluster_std=1.5,
    random_state=42
)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

dbscan = DBSCAN(eps=0.4, min_samples=8)
labels = dbscan.fit_predict(X_scaled)

unique_labels = set(labels)
num_clusters = len([l for l in unique_labels if l != -1])

num_noise = np.sum(labels == -1)

print("Number of clusters detected (excluding noise):", num_clusters)
print("Number of noise points:", num_noise)

```

```python
'''

48. Load the Digits dataset, reduce dimensions using t-SNE, then
apply Agglomerative Clustering and plot the clusters.

'''

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

digits = load_digits()
X = digits.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

agg = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels = agg.fit_predict(X_tsne)

plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, s=35, edgecolor='k')
plt.title("Agglomerative Clustering on Digits Dataset (t-SNE Reduced)")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.tight_layout()
plt.show()
```
